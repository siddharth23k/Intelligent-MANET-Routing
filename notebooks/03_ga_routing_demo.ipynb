# 03 - Genetic Algorithm Routing Demo

This notebook demonstrates GA-based routing using the trained Random Forest model
as a link stability predictor.

It reuses the functions defined in `ml/ga_routing.py` and `ml/data_utils.py`.

Make sure you have already trained and saved the models by running `ml/train_model.py`
or the training notebook before executing this demo.


```python
import sys
import os
sys.path.insert(0, os.path.abspath('..'))
os.chdir('..')
from ml import ga_routing
best_route, best_score, history = ga_routing.run_ga()
print("Best route:", best_route)
print("Route fitness:", best_score)
```

```python
import matplotlib.pyplot as plt

plt.plot(history)
plt.xlabel("Generation")
plt.ylabel("Best fitness")
plt.title("GA convergence over generations")
plt.show()
```
